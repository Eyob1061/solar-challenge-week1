```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('../data/benin-malanville.csv')  # Adjust path if needed
df.head()
```

```python
df.describe()
df.isna().sum()
missing_cols = df.columns[df.isna().mean() > 0.05]
print("Columns with >5% nulls:\n", missing_cols)

```

```python
cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
z_scores = np.abs(zscore(df[cols].dropna()))
outliers = (z_scores > 3).sum(axis=0)
print("Outliers count per column:\n", dict(zip(cols, outliers)))
```

```python
for col in cols:
    if df[col].isna().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
        
```

```python
df.to_csv('../data/benin_clean.csv', index=False)
```

```python
df.to_csv('../data/benin_clean.csv', index=False)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

plt.figure(figsize=(15,5))
for col in ['GHI', 'DNI', 'DHI', 'Tamb']:
    plt.plot(df['Timestamp'], df[col], label=col)
plt.legend()
plt.title("Solar & Temp Time Series")
plt.show()
```

```python
df.groupby('Cleaning')[['ModA', 'ModB']].mean().plot(kind='bar')
plt.title("Sensor Performance: Cleaned vs Not")
plt.show()
```

```python
sns.heatmap(df[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

```

```python
sns.scatterplot(x='WS', y='GHI', data=df)
plt.title("Wind Speed vs GHI")
plt.show()

```

```python
df['WindDir'] = df['WD'] // 30 * 30
sns.histplot(data=df, x='WindDir', bins=12)
plt.title("Wind Direction Histogram")
plt.show()

sns.histplot(df['GHI'], kde=True)
```

```python
sns.scatterplot(x='RH', y='Tamb', data=df)
plt.title("RH vs Temperature")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['GHI'], df['Tamb'], s=df['RH'], alpha=0.5)
plt.xlabel("GHI")
plt.ylabel("Tamb")
plt.title("GHI vs Tamb (Bubble = RH)")
plt.show()
```
